###La mejor manera de abrir archivos

```python
#urllib.request es un módulo Python para acceder y utilizar recursos de internet identificados por URLs (Uniform Resource Locators).
#import urllib.request
```

```python
#abrir archivos con función with ... la mejor práctica
with open("ORD_GREMIALES_202208.txt", "r") as file1:
  FileContent = file1.read()
  print(FileContent)
```

```python
file1.closed
```

```python
print(FileContent)
```

```python
#para leer los primeros n caracteres
with open("ORD_GREMIALES_202208.txt","r") as file1:
  print(file1.read(15))
```

###Escribir un documento con la función **open**

```python
file8 = open("firstfile.txt", "w")
```

```python
#Para escribir en el archivo creado con with:
with open("fileWith.txt","w") as file1:
  file1.write("Línea A\n")
  file1.write("Línea B\n")
```

```python
#para agregar elementos
with open("fileWith.txt","a") as file1:
  file1.write("Linea C\n")
```

```python
#Otra forma de escribir sobre el archivo que se creó
lineas = ["Línea C\n", "Línea D\n"]
with open("firstfile.txt", "w") as file1:
  for line in lineas:
    file1.write(line) 
```

```python
#Copiar el archivo
with open("fileWith.txt","r") as readfile:
  with open("fileCopy.txt","w") as writefile:
    for line in readfile:
      writefile.write(line)
```

```python
#Leer el archivo que se acaba de generar  
with open("fileWith.txt","r") as testwrite:
  print(testwrite.read())
```

```python
with open("fileCopy.txt","w") as writefile:
  writefile.write("Línea D\n")
```

```python
with open("fileCopy.txt","r") as writefile:
  print(writefile.read())
```

###Agregar filas a un archivo
