## Python Packages for Data Science

La librería de **Pandas** de Python especializada en el manejo y análisis de estructuras de datos.

La librería **NumPy** de Python esta especializada en el cálculo numérico y el análisis de datos, especialmente para un gran volumen de datos. Incorpora una nueva clase de objetos llamados arrays que permite representar colecciones de datos de un mismo tipo en varias dimensiones, y funciones muy eficientes para su manipulación.

La librería **SciPy** contiene módulos para optimización, álgebra lineal, integración, interpolación, funciones especiales, FFT, procesamiento de señales e imagen, resolución de EDOs y otras tareas relacionadas con la ciencia e ingeniería.


## Python Packages for Visualization

La librería **Matplotlib** permite crear y personalizar los tipos de gráficos más comunes, como barras, Histogramas, gráficos de líneas .. etc.

La librería **Seaborn** permite generar gráficos de alto nivel para creación de gráficas atractivas como mapas de calor, está basada en la librería de Matplotlib.

## Python Packages for Algorithmic

La librería **Scikit - learn**  es un conjunto de rutinas escritas en Python para hacer análisis predictivo, que incluyen clasificadores, algoritmos de clusterización, etc. Está basada en NumPy, SciPy y matplotlib, de forma que es fácil reaprovechar el código que use estas librerías.

La librería de **Statsmodels** proporciona clases y funciones para la estimación de muchos modelos estadísticos diferentes, así como para realizar pruebas estadísticas y exploración de datos estadísticos.

## Utilizando la librería de Pandas ...

Se pueden importar diferentes tipos de archivos como .csv, .json, .xlsx....

Para importar una base de datos se debe de poner la ruta de acceso ya sea un ruta dentro de la misma computadora o un link de internet

```python
#Importando la librería pandas
import pandas as pd
```

```python
url ='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/Cereals.csv'
df = pd.read_csv(url, header = None)
```

```python
#Si queremos ver los últimos 10 registros de la BD
df.tail(10)
```
